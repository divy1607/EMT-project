## *I. Introduction* :

1) What is Corrugated Surface in electrostatic ?

* In the context of electrostatics, a corrugated surface refers to a surface
   that has a series of ridges and troughs, resembling the shape of waves or corrugations. These surfaces are designed to enhance the efficiency and effectiveness of various electrostatic devices or applications.

2) Introducing corrugated surfaces into electrostatic problems using a         perturbative approach can offer several benefits:  

 *Analytical Insight:* Perturbative methods allow for the systematic analysis of the effects of small perturbations (in this case, the introduction of corrugated surfaces) on the overall electrostatic system. This can provide valuable analytical insights into the behavior of the system, such as changes in electric field distribution, potential energy, or charge distribution.

*Simplicity:* Perturbative methods often simplify complex problems by breaking them down into simpler components. By treating the corrugated surface as a perturbation to an existing electrostatic system with a flat surface, the problem becomes more manageable, facilitating the analysis and understanding of the effects of the corrugations.

*Quantitative Predictions:* Perturbative approaches can provide quantitative predictions of how the corrugated surface alters various electrostatic properties, such as capacitance, charge distribution, or electric field strength. This allows engineers and researchers to assess the impact of corrugations on the performance of electrostatic devices accurately.

*Computational Efficiency:* Perturbative methods can be computationally more efficient compared to numerical simulations, especially for small perturbations. This efficiency enables rapid exploration of different corrugation geometries or parameters, aiding in the optimization of electrostatic devices.

*Physical Insights:* By analyzing the perturbed system, researchers can gain deeper insights into the underlying physics of electrostatic interactions with corrugated surfaces. This understanding can lead to the development of novel electrostatic devices with improved performance and efficiency.5

Smooth surfaces can exhibit corrugations that impact their interactions, even on a small scale. For instance, consider a charge near a surface. The image potential, resulting from induced charge distribution, is crucial for forming image potential states, which are quantum states of electrons localized at the surface of materials with negative electron affinity. Understanding the behavior of these electrons can reveal important physical and chemical surface properties. Prior to 1980, most studies of image potential effects assumed planar interfaces between two media, but the presence of deformations modifies the image potential.

Perturbative expansions are commonly used in electromagnetism courses, particularly in multipolar expansions. In 1985, Clinton et al. employed a perturbative approach to solve Poisson's equation, calculating the image potential for a point charge in vacuum near a perfectly conducting corrugated surface. Recently, this approach was also used to analyze electron-electron interaction in a 2D system with a corrugated dielectric medium and to study van der Waals interaction between an anisotropic polarizable particle and a nonplanar surface.

In this paper, we review the calculations of Clinton et al., aiming to make their results more accessible. We focus on the case of a sinusoidal surface, demonstrating how the potential, electric field, and induced surface charge density are influenced by the corrugated surface. Additionally, we introduce an original contribution by extending this perturbative approach to solve Laplace's equation for the electrostatic potential of a corrugated neutral conducting cylinder in a uniform electric field.

 under the boundary conditions

 $ G(\mathbf{r,r'})|_{z=0}=0, \hspace{20cm} (3)$

$ G(\mathbf{r,r'})|_{|r|>>|r'|}\rightarrow0 . \hspace{20cm} (4)$

## *III. REVIEW OF CLINTON, ESRICK, AND SACKS' CALCULATIONS*

In this section, we review the perturbative analytical calculations with Perturbatively, the potential $\Phi$ for a point charge Q located at *r'* ,in the presence of a grounded conducting corrugated surface given by $ z = h(r_{||})$, where $h(r_{||})$ is a function describing the small corrugation in the surface $[max|h(r_{||}) << z']$ . Effectively, the problem consists of finding [in the region $z \geq h(r_{||})$] the solution of Eq ${\nabla^2}G(r,r') = -4{\pi}\delta(r - r')$ .Under the condition of eq $G(r, r')|{(|r| >>|r'|)} ⟶ 0$, but with the boundary condition $G(r, r')|_{z=0} = 0$, modified to $G(r, r')|_{z=hr_{||}} = 0\hspace{20cm}(6)$

(Fig 3. Illustration of charge q located at $r' = r_{||}+z'\hat{z}$ with (z'>0) interacting with grounded conducting corrugated surface, whose corrugation is described by $z = h(r_{||}).$)

To build a perturbative description, these authors introduced an arbitrary auxiliary parameter $\epsilon$, with $0<\epsilon \le 1$. Thus, one parameterrizes: $h(r_{||}) \longrightarrow \varepsilon h(r_{||})$,so that Eq (6) is replaced by

$G(r, r')|_{z = \varepsilon h(r_{||})} = 0 \hspace{20cm} (7)$,

and Green's function is written as a perturbative expansion in terms of the parameter $\epsilon$,

$G(r,r') = G^{(0)}(r, r') +  \sum_{n=1}^{\infty} \epsilon ^{n}G^{n}(r, r') \hspace{20cm} (8)$


where $G^{(0)}(r, r')$ is the unperturbed Green's function[Eq(5), Now considered for $z \ge h(r_{||})$], and $G^{(n)}(r, r')$ are perturbative corrections. Where $\epsilon = 0$, one  has the solution to the case of a planar surface, whereas, when $\epsilon = 1$,  we recover the actual problem of a corrugated surface described by $z = h(r_{||})$

Substituting Eq. (8) in Eq. (2), one obtains
$\nabla ^{2}G^{(0)}(r,r') + 4\pi\delta (r-r') +  \sum_{n=1}^{\infty} \epsilon ^{n}[\nabla^{2}G^{(n)}(r,r')]= 0 \hspace{20cm} (9)$


Since $\epsilon$ is arbitrary, the coefficints multiplying each power of $\epsilon$ must vanish. In this way, one obtains the differential equations for each $G^{(n)}$, which are given by

$\nabla ^{2}G^{(0)}(r, r')=-4\pi\delta (r-r') \hspace{20cm} (10)$  

$\nabla^{2}G^{(n)}(r,r')=0 $  $(n\ge 1) \hspace{20cm} (11)$

Using Eq. (8), the boundary condition Eq. (7) becomes

$\sum_{i=0}^{\infty}\varepsilon^{i}G^{(i)}(r,r')|_{z=\varepsilon h(r_{||})}= 0 \hspace{20cm} (12)$

Expanding it in power of $\varepsilon h_{(r_{||})}$, one has
$\sum_{i=0}^{\infty} \sum_{m=0}^{\infty} \frac{e^{i+m} h(r_{||})^{m}}{m!} \frac{\partial^{m}}{\partial z^{m}} G^{(i)}(r, r') \bigg|_{z=0} = 0 \hspace{20cm} (13)
$

Grouping together terms which are same order in $\epsilon$, one obtains

$G^{(0)}(r, r') \bigg|_{z=0} + [\varepsilon \frac{\partial}{\partial z}h(r_{||}) \bigg|_{z=0} \frac{\partial}{\partial z}G^{(0)}(r, r') \bigg|_{z=0} + G^{(1)}(r, r') \bigg|_{z=0}] + [\epsilon ^2 \frac{1}{2} \frac{\partial^2}{\partial z^2}h(r_{||}) \bigg|_{z=0} \frac{\partial^2}{\partial z^2}G^{(0)}(r, r') \bigg|_{z=0} + h(r_{||}) \frac{\partial}{\partial z}G^{(1)}(r, r') \bigg|_{z=0} + h(r{||}) \frac{\partial}{\partial z}G^{(2)}(r, r') \bigg|_{z=0}] +...=0 \hspace{20cm} (14)$    

Again, $\varepsilon$ is arbitrary, so solving Eq(14) requires that the coefficients multiplying each power of $\epsilon$ vanish.In this way, one obtains boundary conditions at z=0 for $G^{(0)} and G^{(n)}$,

$G^{(0)}(r,r')|_{z=0}=0$

$ G^{(1)}(r, r') \bigg|_{z=0}=-\frac{\partial}{\partial z}h(r_{||}) \bigg|_{z=0} \frac{\partial}{\partial z}G^{(0)}(r, r') \bigg|_{z=0} $

$ G^{(2)}(r, r') \bigg|_{z=0}=-\frac{1}{2} \frac{\partial^2}{\partial z^2}h(r_{||}) \bigg|_{z=0} \frac{\partial^2}{\partial z^2}G^{(0)}(r, r') \bigg|_{z=0} - h(r_{||}) \frac{\partial}{\partial z}G^{(1)}(r, r') \bigg|_{z=0} $

We can put all these equation together as follows:
$G^{(0)}(r,r')|_{z=0 = 0} \hspace{20cm} (15)$,

$G^{(n)}(r,r')|_{z=0} = - \sum{m=1}^{n} \frac{ h(r_{||})^{m}}{m!} \frac{\partial^{m}}{\partial z^{m}} G^{(n-m)}(r, r') \bigg|_{z=0} (n\ge1) \hspace{20cm} (16)$  

It is important to remark that only by knowing $G^{(0)}$, one can obtain the boundary condition for $G^{(1)}$, and then for $G^{(2)}$ and so on. Also note that for $n \ge 1$, these boundary conditions are inhomogeneous. In other words, the original problem of solving the non-homogeneous Eq. (2), under a homogeneous boundary condition on a complicated surface $z = h(r_{||})$[Eq.(6)] is now replaced by the problem of solving the homogeneous Eq. (11), under non-homogeneous boundary conditions on a simple plane at z = 0 [Eq. (16)]. Following Ref. 8, in order to find a solution, it is convenient to introduce the Fourier transformation (and its inverse)

$\tilde{f}(k,r'_{||};z,z') = \int d^2r_{||} \, e^{ikr} f(r,r') \hspace{20cm} (17)$   

$f(r,r') = \frac{1}{2\pi}^{2} \int d^2k \frac{e^{ik \cdot r_{||}}}{|k|^2} \tilde{f}(k,r_{||};z,z')
\hspace{20cm} (18)
$.

Writing $G(r,r'), G^{(0)}(r,r'), and G^{(n)}(r,r')$ as inverse Fourier transformations and making use of the fact that the Dirac delta function is given in Fourier space as

$\delta(r-r') = \frac{1}{(2\pi)^2} \int d^2k \, e^{ik \cdot (r_{||} - r'_{||})}\delta(z-z') \hspace{20cm} (19)
$

we can write Eqs.(10) and (11) as

$(\frac{\partial^2}{\partial z^2} -|k|^2) \tilde{G}^{(0)}(k, r'_{||}; z, z') = -4\pi e^{-ik \cdot r'{||}} \delta(z-z')  \hspace{20cm} (20)$

$(\frac{\partial^2}{\partial z^2} -|k|^2) \tilde{G}^{(n)}(k, r'_{||};  z, z') = 0, (n \ge 1) \hspace{20cm} (21)$

One can also find the boundary conditions given in Eqs. (15) and (16) in the Fourier space

$\tilde{G^{(0)}}(k,r'_{||};z,z')|_{z=0}=0 \hspace{20cm} (22)$

$\tilde{G^{(n)}}(k, r'_{||};  z, z') \bigg|_{z=0} = -\sum_{m=1}^{n} \frac{1}{m!} \int d^2k' \frac{1}{(2\pi)^2} \tilde{h}m(k- k') \frac{\partial^m}{\partial z^m} \tilde{G}^{(n-m)}(k', r'_{||};  z, z') \bigg|_{z=0}
\hspace{20cm} (23)
$  

where $\tilde{h}m(k- k')= \int d^2r_{||} \, e^{-i(k - k') \cdot r_{||}} [h(r_{||})]^{m} \hspace{20cm} (24)
$

Under the boundary condition Eq. (22) and the condition that $\tilde{G}^{(0)}$ goes to zero zero for large distances [Eq. (4)], the solution of Eq. (20) is

$\tilde{G}^{(0)}(k, r'_{||}; z, z') = \frac{2\pi}{|k|} e^{-ik \cdot r'_{||}}[ e^{-|k||z - z'|} - e^{-|k|(z+z')]}\hspace{20cm} (25)$

By performing an inverse Fourier transformation on $\tilde{G}^{(0)}$, one obtains the unperturbed Green's function which exactly matches.

Under the boundary condition Eq. (23) and the condition that $\tilde{G}^{(n)}$ goes to zero for large distances, the solution of Eq(21) is

$\tilde{G}^{(n)}(k, r'_{||}; z, z') = -e^{-|k|z} + \sum{n=1}^{\infty} \sum_{m=1}^{n} \frac{1}{m!} \int d^2k' \frac{1}{(2\pi)^2} \tilde{h}m(k - k') \frac{\partial^m}{\partial z^m} \tilde{G}^{(n-m)}(k', r'_{||}; k, z; z') \bigg|_{z=0}
\hspace{20cm} (26)
$

Thus, one obtains the functions $\tilde{G}^{(n)}$ recursively in terms of $\tilde{G}^{(0)}$. The solution for $G^{(n)}$ is obtained by performing the inverse Fourier transformation,

$G^{(n)}(r, r') = \frac{1}{2\pi} \int d^2k \, e^{ik \cdot r_{||}} \tilde{G}^{(n)}(k, r'_{||};  z, z')
\hspace{20cm} (27)
$  

Thus, substituting $G^{(0)}$ [Eq(5)] and $G^{||}$ [Eqs(26) and (27)] in Eq(8) (with $\epsilon=1$), one obtains Green's function for the problem of a point charge in the presence of a grounded conducting corrugated surface.

##  *V. CONDUCTING CYLINDER IN A UNIFORM ELECTRIC FIELD*

In this section, we review the textbook solution to the problem of a cylinder in the presence of a uniform electrostatic field, and in Sec. VI, we introduce corrugation to the cylinder's surface.


We start by considering an infinite grounded conducting cylinder, with radius a, in a uniform electrostatic field $E_{0} = E_{0} \hat{x}$, as illustrated in Fig. The electrostatic potential can be calculated by means of Laplace's equation $\nabla^{2}\Phi(r,r')=0$. Since the cylinder is infinitely long, we have translational symmetry in the z-direction, so that $\Phi{(\rho, \theta, z)} \longrightarrow \Phi{(\rho, \theta)}$. In this case, Laplace's equation in cylindrical cooordinates can be written as

$\frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial \Phi(\rho,\theta)}{\partial \rho} \right) + \frac{1}{\rho^{2}} \frac{\partial^2 \Phi(\rho,\theta)}{\partial \theta^{2}} + \frac{1}{\rho^{2}} \frac{\partial^2 \rho(\rho,\theta)}{\partial \rho^{2}} = 0
\hspace{20cm} (31)
$

![picture](https://drive.google.com/uc?export=view&id=1cfXEcRXd2IH8Ti5hARZJBi6x4nTT3HnU)

(Fig 6. Cross section of an infinite grounded conducting cylinder, with radius a, placed in a uniform electrostatic field, $\mathbf{E_0} = E_0\hat{x}$)

By  using the method of separation of variables, we obtain

$\Phi{(\rho, \theta)} = \sum_{m=1}^{\infty}(A_{m}\rho^{m}+B_{m}\rho^{-m})\times[C_{m}cos(m\theta) + D_{m}sin(m\theta)] \hspace{20cm} (32)$

Since we consider the cylinder as a grounded perfect conductor, we have a boundary condition at $\rho=a$,

$\Phi{(\rho,\theta)=0} \hspace{20cm} (33)$.

At large $\rho$, the potential has to describe the field $E_{0}=E_{0}\hat{x}$ and, Theefore, it can be written as

 $\Phi(\rho,\theta)|_{\rho >>a} \longrightarrow -E_{0}\rho cos(\theta) \hspace{20cm} (34)$

 Thus, from these boundary conditions, one obtains

 $\Phi{(\rho, \theta)}= -E_{0}\rho cos(\theta)(1-\frac{a^{2}}{\rho^{2}}) \hspace{20cm} (35)$


outside the cylinder, while inside the cylinder the potential is zero

##*VI. CONDUCTING CORRUGATED CYLINDER PLACED IN A UNIFORM ELECTRIC FIELD*#

We now generalize Sec. V to the case of an infinite grounded conducting parametrized by $\rho(\theta)=a+h(\theta),$ where $|h(\theta)|$ is a small corrugation, as illustrated in Fig. 7. The potential can be calculated from Eq. (31), under the condition that the
potential describes the field $\mathbf{E_0} = E_0\mathbf{\bar{x}}$ for large distances [Eq. (34)].

Following the perturbative approach discussed in Sec. III,
we introduce an arbitrary auxiliary parameter $\varepsilon$, with $0\le\varepsilon\le1$, and replace $h(\theta)\rightarrow \varepsilon h(\theta),$ so that Eq.(33) is replaced by

$\Phi(\rho, \theta)|_{\rho = a+\varepsilon h(\theta)}=0,\hspace{20cm} (36)$

and the electrostatic potential is written as a perturbative expansion in $\varepsilon$ as

$\Phi(\rho, \theta) = \Phi^{(0)}(\rho, \theta)+\sum_{n=1}^{\infty}\varepsilon^n \Phi^{(n)}(\rho, \theta). (37) \hspace{20cm}$

When $\varepsilon=0$, one has the non corrugated cylinder $(\rho= a),$ whereas, when $\varepsilon=1$, we recover the actual corrugated cylinder $[\rho=  a+h(\theta)].$ Substituting Eq. (37) in Eq. (31), one
has

$\sum_{n=0}^{\infty} \varepsilon^n[\frac{1}{\rho} \frac{\partial}{\partial\rho}(\rho \frac{\partial}{\partial\rho})+\frac{1}{\rho^2}\frac{\partial^2}{\partial\theta^2}]\Phi^{(n)}(\rho, \theta)=0  \hspace{20cm} (38)$

Since $\varepsilon$ is arbitrary, the coefficients multiplying each power of $\varepsilon$ must vanish. Then, one has

$[\frac{1}{\rho}\frac{\partial}{\partial\rho}(\rho \frac{\partial}{\partial\rho})+\frac{1}{\rho^2}\frac{\partial^2}{\partial\theta^2}]\Phi^{(n)}(\rho, 0)=0. \hspace{1cm} (n\ge0) \hspace{20cm} (39)$

We can find the boundary conditions for Eq. (39) by
applying the same procedure discussed in Sec. III. Using
Eq. (36) in Eq. (37) and expanding in powers of $\varepsilon h(\theta), $ one has

$\sum_{i=0}^{\infty} \sum_{m=0}^{\infty} \varepsilon^{i+m}[h(\theta)]^m \frac{1}{m!} \frac{\partial^m}{\partial\rho^m} \Phi^{(i)} (\rho, \theta)|_{\rho=a}=0. \hspace{20cm}(40)$

Collecting terms of the same order in $\varepsilon, $ one obtains

$\Phi^{(0)}(\rho, \theta)|_{\rho=a}+\varepsilon[h(\theta)\frac{\partial}{\partial\rho}(\rho, \theta)|_{\rho=a}+\Phi^{(1)}|_{\rho=a}] + \varepsilon^2[h(\theta)^2 \frac{1}{2} \frac{\partial^2}{\partial\rho^2}\Phi^{(0)}(\rho, \theta)|_{\rho=a}+h(\theta)\frac{\partial}{\partial\rho}\Phi^{(1)}(\rho, \theta)|_{\rho=a}+\Phi^{(2)}(\rho, \theta)|_{\rho=a}]+...=0 \hspace{20cm} (41)$

Once again, the coefficients multiplying each power of $\varepsilon$ must vanish. In this way, we obtain boundary conditions

$\Phi^{(0)}(\rho, \theta)|_{\rho=a}=0, \hspace{20cm} (42)$

$\Phi^{(n)}(\rho, \theta)|_{\rho=a} = -\sum_{m=1}^{n} \frac{[h(\theta)]^m}{m!} \frac{\partial^m}{\partial\rho^m} \Phi^{(n-m)}(\rho, \theta)|_{\rho=a} \hspace{1cm} (n\ge1). \hspace{20cm} (43)$

We also have following conditions:

$\Phi^{(0)}(\rho, \theta)|_{\rho>>a} \rightarrow -E_0 \rho cos\theta, \hspace{20cm} (44)$

$\Phi^{(n)}(\rho, \theta)|_{\rho>>a} \rightarrow 0. \hspace{20cm} (45)$

These conditions come from the fact that sufficiently far
from the cylinder $(\rho>>a)$ its influence  is negigible, so that
one recovers the external field $E_0$. Since $\Phi^{(0)}$ already specifies that when $\rho>>a$ the potential is the one corresponding
to the external field, $\Phi^{(n)}$ [Eq. (45)] cannot contribute to the
potential in this limit.

The solution of $\Phi^{(0)}$ is given in Eq. (35). Moreover, the
solution of Eq. (39) can be found by using the method of separation of variables

$\Phi^{(n)}(\rho, \theta) = \sum_{l=1}^{\infty}(A_l \rho^l+B_l \rho^{-l}) \times [C_lcos(l\theta)+ D_lsin(l\theta)]. \hspace{20cm} (46)$

From Eq. (45), one can conclude that $A_l=0.$ Thus, Eq. (46)
becomes

$\Phi^{(n)}(\rho, \theta) = \sum_{l=1}^{\infty}\rho^{-l} [C_lcos(l\theta)+ D_lsin(l\theta)]. \hspace{20cm} (47)$

where we made $B_lC_l=C'_l, $  and $B_lD_l = D'_l.$ Using the boundary condition Eq. (43) in Eq. (47), one has

$\Phi^{(n)}(a, \theta) = \sum_{l=1}^{\infty}a^{-l} [C_lcos(l\theta)+ D_lsin(l\theta)]. \hspace{20cm} (48)$

Multiplying both sides by $cos(p\theta)$ or $sin(p\theta)$ and integrating in $\theta$ from $0$ to $2\pi,$ one obtains

$C'_l = \frac{a'}{\pi}\int_{0}^{2\pi}d\theta cos(l\theta) \Phi^{(n)}(a, \theta), \hspace{20cm} (49)$

$D'_l = \frac{a'}{\pi}\int_{0}^{2\pi}d\theta sin(l\theta) \Phi^{(n)}(a, \theta).
\hspace{20cm} (50)$

Using Eqs. (49) and (50) in Eq. (47), one finds

$\Phi^{(n)}(\rho, \theta)= -\sum_{m=1}^{n} \sum_{l=1}^{\infty}\frac{1}{\pi} (\frac{a}{\rho})^l \times \int_{0}^{2\pi} d\bar{\theta} cos[l(\bar{\theta}-\theta)] \times \frac{[h(\bar{\theta})]^m}{m!} \frac{\partial^m}{\partial\rho^m} \Phi^{(n-m)}(\rho, \theta)|_{\rho=a}. \hspace{20cm} (51)$

Using Eqs. (49) and (50) in Eq. (47), one finds $\varepsilon=1,$ we find the solution for the potential l related to the
problem of a grounded conducting corrugated cylinder
placed in a uniform electric field.


We can write explicitly the perturbative corrections $\Phi^{(n)} up to the second order$

$\Phi^{(1)}(\rho, \theta)=\sum_{l=1}^{\infty} \frac{2E_0\rho}{\pi} (\frac{a}{\rho})^{l+1} \int_{0}^{2\pi} d\bar{\theta_1} cos[l(\bar{\theta_1}-\theta)] \times cos(\bar{\theta_1}) \frac{h(\bar{\theta_1})}{a}, \hspace{20cm} (52)$

$\Phi^{(2)}(\rho, \theta) = \sum_{l=1}^{\infty}\frac{E_0\rho}{\pi} (\frac{a}{\rho})^{l+1} \int_{0}^{2\pi} d\bar{\theta_2} cos[l(\bar{\theta_2}-\theta)] \times [\frac{2l}{\pi} \int_{0}^{2\pi}d\bar{\theta_1} cos [l(\bar{\theta_1}-\bar{\theta_2})] \times cos (\bar{\theta_1})\frac{h(\bar{\theta_2})h(\bar{\theta_1})}{a^2}-cos (\bar{\theta_2}) \frac{h(\bar{\theta_2})^2}{a^2}]. \hspace{20cm} (53)$

In these terms, one can identify the ratio $|h|/a$ as the perturbative parameter so $\Phi^{(n)} \propto (|h|/a)^{n}.$ Thus, as the ratio $|h|/a$ decreases, the faster our  perturbative solution converges. In
addition, the ratio $a/\rho$ controls the convergence of the series in $l$, which means that as the ratio $a/\rho$ decreases, the faster these series converge. Finally, it is important to remark that the solution found here is valid for any function $h(\theta)$.In Secs. VII and VIII, we consider some applications of our results with specific choices for $h(\theta).$

## *VII. CYLINDER WITH A SLIGHT ENHANCEMENT IN ITS RADIUS*

We check the validity of our calculations by choosing a constant corrugation profile $h(\theta)=δ$ which should give the solution for a cylinder with radius $a+δ$.Using Eq.(51) ,for this case we get

$\Phi^{(1)}(\rho, \theta) = 2 \delta E_0 \left(\frac{a}{\rho}\right) \cos(\theta)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(54)$

$\Phi^{(2)}(\rho, \theta) = \frac{\delta^2}{\rho} E_0 \cos(\theta)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\:\;\;(55)$

$\Phi^{(i)}(\rho, \theta) = 0 \quad \text{for} \quad (i \ge 3)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(56)$

Therefore, Eq. (37) (with $ϵ=1$
⁠) can be written as

$\Phi(\rho, \theta) = - E_0 \rho \cos(\theta) \left(1 - \frac{(a + \delta)^2}{\rho^2}\right)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(57)$

which, as expected, corresponds to the result for a cylinder of radius $a+δ$

## *VIII. SINUSOIDAL CORRUGATED CYLINDER*

Let us consider a sinusoidal corrugated cylinder whose profile is

$h(\theta) = \frac{\delta}{2} \left[1 + \cos(\nu \theta)\right]\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(58)$

where $ν\: ϵ\: N$.Note that, when $ν=0$, Eq. (58) describes the case of a cylinder with a slight enhancement in its radius.
In this case, we calculate an approximate solution for $Φ$ up to first-order in h, which is given by

$\Phi(\rho, \theta) \approx \Phi^{(0)}(\rho, \theta) + \Phi^{(1)}(\rho, \theta)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(59)$

with $Φ^{(0)}$ given in Eq. (35). To calculate $Φ^{(1)}$ we use Eq. (58) in Eq. (52) and obtain
$\Phi^{(1)}(\rho, \theta) = \frac{E_0 \delta}{\pi} \sum_{l=1}^{\infty} \left(\frac{a}{\rho}\right)^l \times \int_{0}^{2\pi} d\hat{\theta} \, \cos \left[l(\hat{\theta} - \theta)\right]× \left[1 + \cos\left(\nu (\hat{\theta})\right)\right] \cos\left(\hat{\theta}\right)\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(60)$

The above integral is given by

$\int_{0}^{2\pi} d\hat{\theta} \, \cos \left[l(\hat{\theta} - \theta)\right]× \left[1 + \cos\left(\nu (\hat{\theta})\right)\right] \cos\left(\hat{\theta}\right)$

$=πcos(lθ)δ_{l1}+\frac{π}{2}cos(lθ)(δ_{l,|1-\nu|}+δ_{l,ν+1})\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(61)$

From this, we find that $Φ^{(1)}$is given by

$
  Φ^{(1)}(\rho,θ)=\frac{δE_0a}{ρ}cos(θ)+\frac{δE_0}{2}\left\{(\frac{a}{ρ})^{ν+1}cos[(ν+1)θ]+(\frac{a}{ρ})^{|ν-1|}cos[|ν-1|θ]-δ_{ν1}\right\}\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(62)
$

For $ν=0$, we recover Eq. (54), as expected. In the figure below, we plot the potential for the case of a non-corrugated cylinder and compare it with the case of a corrugated one. The magnitude of the potential for a corrugated cylinder oscillates close to that of a non-corrugated one. As $\frac{ρ}{a}$
